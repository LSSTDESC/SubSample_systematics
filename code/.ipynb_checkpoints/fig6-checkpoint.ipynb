```python
a = np.linspace(0,1,40)
b = np.linspace(0,1,40)
ab = np.meshgrid(a,b)
```

```python
def get_mask(cen_x = 0.2, cen_y = 0.2, radius = 0.1):
    cd_cen = (cen_x,cen_y)
    radius = radius
    mask = ((ab[0] - cd_cen[0])**2+(ab[1] - cd_cen[1])**2)<radius
    return mask
mask = get_mask(0.2,0.2,0.01)
mask2 = get_mask(0.8,0.1,0.02)
mask3 = get_mask(0.1,0.8,0.02)

mask4 = get_mask(0.8,0.8,0.03)
mask5 = get_mask(0.5,0.5,0.02)
mask6 = get_mask(0.1,0.5,0.02)
```

```python
plt.figure(figsize = (8,8),frameon=False)

p = plt.plot(ab[0][mask],ab[1][mask],'r.')
p = plt.plot(ab[0][mask2],ab[1][mask2],'r.')
p = plt.plot(ab[0][mask3],ab[1][mask3],'r.')
plt.plot([],[],'ro', label = r'$R_{1}$')

p = plt.plot(ab[0][mask4],ab[1][mask4],'b.')
p = plt.plot(ab[0][mask5],ab[1][mask5],'b.')
p = plt.plot(ab[0][mask6],ab[1][mask6],'b.')
plt.plot([],[],'bo', label = r'$R_{2}$')

mask_blk = ~mask & ~mask2 & ~mask3 & ~mask4 & ~mask5 & ~mask6
p = plt.plot(ab[0][mask_blk],ab[1][mask_blk],'k,')
plt.plot([],[],'ko', label = r'$R_{3}$')

plt.legend(fontsize = 25)
#plt.text(0.2,0.2, r'$R_{1}$',size = 15)

plt.gca().set_xticks([])
plt.gca().set_yticks([])
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from camb import model, initialpower, CAMBparams, get_results

# Function to compute matter power spectrum
def compute_matter_power_spectrum(fnl=0, b1=2.5, redshift=1.5):
    """
    Computes the matter power spectrum with f_NL corrections.
    
    Parameters:
        fnl (float): Non-Gaussianity parameter (local-type)
        b1 (float): Linear galaxy bias
        redshift (float): Redshift at which to compute the power spectrum
        
    Returns:
        k (array): Wavenumbers in h/Mpc
        P_gal (array): Galaxy power spectrum with f_NL correction
        P_matter (array): Linear matter power spectrum
    """
    # Set up CAMB parameters
    pars = CAMBparams()
    pars.set_cosmology(H0=70, ombh2=0.022, omch2=0.12)
    pars.InitPower.set_params(As=2.1e-9, ns=0.965)
    pars.set_matter_power(redshifts=[redshift], kmax=10)
    results = get_results(pars)
    
    # Get the linear matter power spectrum
    k, z, P_matter = results.get_matter_power_spectrum(minkh=1e-3, maxkh=1e-1, npoints=50)
    #P_matter = P_matter.flatten()  # Ensures it's a 1D array
    P_matter = np.squeeze(P_matter)  # Remove any extra dimensions





    # Compute the f_NL-induced scale-dependent bias correction
    delta_c = 1.686  # Critical overdensity for collapse
    Omega_m = pars.omegam  # Matter density today
    H0 = pars.H0   # H0 in units of 100 km/s/Mpc
    #D_z = results.get_redshift_evolution(k, [redshift], ["delta_cdm"])[0]  # Growth function
    D_z = np.squeeze(results.get_redshift_evolution(k, [redshift], ["delta_cdm"])[0])
    #T_k = results.get_matter_transfer_data().transfer_z('delta_cdm')[0]  # Matter transfer function
    T_k = np.squeeze(results.get_matter_transfer_data().transfer_z('delta_cdm')[0])
    
    # Compute the scale-dependent bias correction
    bias_correction = (b1 - 1) * (2 * fnl * delta_c * Omega_m * H0**2) / ((3*10**5)**2*(k**2))
    print(bias_correction)
    # Compute the galaxy power spectrum with f_NL correction
    P_gal = (b1 + bias_correction) ** 2 * P_matter

    return k, P_gal, P_matter

# Run the function for different f_NL values
fnl_values = [0,20]
b1 = 2  # Example galaxy bias
z = 1   # Redshift

plt.figure(figsize=(8, 6))

for fnl in fnl_values:
    k, P_gal, P_matter = compute_matter_power_spectrum(fnl, b1, z)
    plt.loglog(k, P_gal, label=f"f_NL = {fnl}")

# Plot settings
plt.xlabel(r"$k$ [h/Mpc]")
plt.ylabel(r"$P_g(k)$ [Mpc/h]^3")
plt.legend()
plt.title("Galaxy Power Spectrum with Non-Gaussianity")
plt.grid(True, which="both", linestyle="--", alpha=0.5)
plt.show()
```

```python
from pyccl import Cosmology as cosmologyccl
import numpy as np
Omega0_cdm=0.26377065934278865
Omega0_b=0.0482754208891869
h=0.67556
sigma8=0.8225
n_s=0.9667
redshift = 0.0
delta_crit = 1.686      # critical overdensity for collapse from Press-Schechter
c = 299792 #km/s
cosmoccl = cosmologyccl(Omega_c = Omega0_cdm, Omega_b = Omega0_b, h = h, sigma8 = sigma8, n_s = n_s)
H0 = cosmoccl['H0']
DH = (H0/c) # in units of h/Mpc, eq. 4 https://arxiv.org/pdf/astro-ph/9905116.pdf
Omega0_m = Omega0_cdm+Omega0_b
k_ = np.logspace(-4, 10, 10000)
Dc = cosmologyccl.comoving_radial_distance
Dlin = cosmoccl.growth_factor  
```

```python
import camb
from scipy.interpolate import InterpolatedUnivariateSpline as IUS
default_sigma8 = 0.79636317
default_As = 2e-9
P_k_max=10.0
scale = (sigma8/default_sigma8/1.0005116352583587)**2
As = scale*default_As
pars = camb.set_params(H0=100*h, ombh2=Omega0_b*h*h, omch2=Omega0_cdm*h*h, mnu=0.06, omk=0, tau=0.06,
                    ns=n_s, halofit_version='mead', lmax=3000,WantTransfer=True)
pars.InitPower.set_params(ns=n_s, As  = As)
pars.set_matter_power(redshifts=[0], kmax=P_k_max)
results= camb.get_results(pars)
```

```python
kh,zs,PK = results.get_matter_power_spectrum(maxkh=10.0)
Plin_ = IUS(kh, PK[0,:], ext=1)
Plin = IUS(k_, Plin_(k_), ext=1)

trans = results.get_matter_transfer_data()
kh = trans.transfer_data[0,:,0]
trans_ = trans.transfer_data[model.Transfer_cdm-1,:,0]
Tlin_ = IUS(kh, trans_/trans_.max(), ext=1)
tk_ = Tlin_(k_)
is_good = np.isfinite(tk_)
Tlin = IUS(k_[is_good], tk_[is_good], ext=1)

```

```python
kh = np.logspace(-3,-1)
#plt.loglog(kh, Plin(kh))

alpha_fnl = 3.0*delta_crit*Omega0_m*(DH*DH)
delta_b = alpha_fnl/(kh**2*Tlin(kh)*Dlin(a = 0.8))
fnl = 10
fnl2 = 5
fnl3 = 0
delta_b = (1+fnl*alpha_fnl/(kh**2*Tlin(kh)*Dlin(a=0.8)))
delta_b2 = (1+fnl2*alpha_fnl/(kh**2*Tlin(kh)*Dlin(a=0.8)))
delta_b3 = (1+fnl3*alpha_fnl/(kh**2*Tlin(kh)*Dlin(a=0.8)))
plt.loglog(kh, delta_b*Plin(kh), label = "fnl = 10")
plt.loglog(kh, delta_b2*Plin(kh), label = "fnl = 5")
plt.loglog(kh, delta_b3*Plin(kh), label = "fnl = 0")
plt.legend()
```
