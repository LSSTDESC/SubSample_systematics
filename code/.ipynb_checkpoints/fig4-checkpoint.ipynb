```python
import pyccl as ccl
import numpy as np
import matplotlib.pyplot as plt
ells = np.arange(1000)
cosmo = ccl.Cosmology(Omega_c=0.26377065934278865, Omega_b=0.0482754208891869,
                          h=0.67556, n_s=0.9667, sigma8=0.8225, transfer_function = "boltzmann_camb", m_nu=0.0)
```

```python
zs = np.linspace(0.1,2,1000)

plt.figure(figsize=(8,6))
sigma = 0.15
dndz = np.exp(-(zs-0.8)**2/sigma**2)
dndz2 = np.exp(-(zs-0.8+0.05)**2/sigma**2)
dndz3 = np.exp(-(zs-0.8-0.05)**2/sigma**2)
lens = np.loadtxt("../data/lens_nz.txt")
source = np.loadtxt("../data/source_nz.txt")

plt.plot(zs, dndz, label = r'$\Delta E(B-V) = 0$',color = 'C1',linewidth = 2)
plt.plot(zs, dndz2, label = r'$\Delta E(B-V) = \pm 0.01$',color = 'C1',linewidth = 2,linestyle = '--')
plt.plot(zs, dndz3, color = 'C1',linewidth = 2,linestyle = '--')
plt.fill_between(lens[0], lens[1]/lens[1].max(), np.zeros_like(lens[0]),label = "whole sample",color = 'C1',linewidth = 2, alpha = 0.3)

#plt.plot(real_nz[:,0],real_nz[:,1]/real_nz[:,1].max(),label = 'Whole Sample',linewidth = 2, color = 'C0')
plt.legend(fontsize = 13)
plt.gca().set_xlim((0.2,2))
plt.xlabel('redshift',size = 15)
p=plt.xticks(size=16)
p=plt.yticks(size=16)

zs2 = np.linspace(0.1 ,2.2,1000)
bias = 1.4*np.ones_like(zs2)
dndz = np.exp(-(zs2-0.8-0.5)**2/sigma**2)
dndz2 = np.exp(-(zs2-0.8-0.5+0.05)**2/sigma**2)
dndz3 = np.exp(-(zs2-0.8-0.5-0.05)**2/sigma**2)

#plt.plot(zs2, dndz, label = r'$\Delta E(B-V) = 0$',color = 'C2',linewidth = 2)
#plt.plot(zs2, dndz2, label = r'$\Delta E(B-V) = 0.01$',color = 'C2',linewidth = 2,linestyle = '--')
#plt.fill_between(source[0], source[1]/source[1].max(), np.zeros_like(source[0]),label = "whole sample",color = 'C2',linewidth = 2, alpha = 0.3)
plt.plot(zs2, dndz, color = 'C2',linewidth = 2)
plt.plot(zs2, dndz2, color = 'C2',linewidth = 2,linestyle = '--')
plt.plot(zs2, dndz3, color = 'C2',linewidth = 2,linestyle = '--')
plt.fill_between(source[0], source[1]/source[1].max(), np.zeros_like(source[0]),color = 'C2',linewidth = 2, alpha = 0.3)


#plt.plot(np.linspace(0.2,0.7,10), np.zeros(10),color = 'C2', linewidth=2)
#plt.plot(real_nz[:,0],real_nz[:,1]/real_nz[:,1].max(),label = 'Whole Sample',linewidth = 2, color = 'C0')
plt.legend(fontsize = 13,loc=1)
plt.gca().set_xlim((0.2,2))
plt.xlabel('redshift',size = 15)
p=plt.xticks(size=16)
p=plt.yticks(size=16)

plt.text(0.3,0.8,'lens galaxy',size = 15,color = 'C1')
plt.text(1.5,0.4,'source galaxy',size = 15,color = 'C2')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fn_ng = np.loadtxt("../data/wtheta_measured_res2048_NG.txt")
fn_gg_pl = np.loadtxt("../data/wtheta_measured_res2048_GG+.txt")
fn_gg_mn = np.loadtxt("../data/wtheta_measured_res2048_GG-.txt")

fn_ng_real = np.loadtxt("../data/wthta_ng.txt").transpose()
fn_gg_pl_real = np.loadtxt("../data/wthta_gg+.txt").transpose()
fn_gg_mn_real = np.loadtxt("../data/wthta_gg-.txt").transpose()
```

```python
from matplotlib.ticker import FuncFormatter
plt.figure(figsize = (7,5))
plt.plot(fn_ng[0],1e4*fn_ng[0]*fn_ng[1],linewidth = 3, label = "sub-sample systematics",color = "C1")
plt.plot(fn_ng_real[0],1e4*fn_ng_real[0]*fn_ng_real[1],linestyle = ':',linewidth = 3, label = "clean", color = "C0")

#plt.gca().set_yscale('log')
plt.gca().set_xscale('log')
plt.ylabel(r'$10^4\cdot\theta \cdot \gamma(\theta)$',size = 18)
plt.xlabel(r'$\theta (degree)$',size = 18)
p1 = plt.xticks(size = 18)
p2 = plt.yticks(size = 18)
plt.legend(fontsize = 18)

plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
```

```python
plt.figure(figsize = (7,5))
plt.plot(fn_gg_pl[0][:170],1e5*fn_gg_pl[0][:170]*fn_gg_pl[1][:170],linewidth = 3, label = "sub-sample systematics",color = "C1")
plt.plot(fn_gg_pl_real[0][::5][:170],1e5*fn_gg_pl_real[0][::5][:170]*fn_gg_pl_real[1][::5][:170],linestyle = ':',linewidth = 3, label = "clean", color = "C0")
#plt.gca().set_yscale('log')
plt.gca().set_xscale('log')
plt.ylabel(r'$10^5\cdot\theta \cdot \xi_{+}(\theta)$',size = 18)
plt.xlabel(r'$\theta (degree)$',size = 18)
p1 = plt.xticks(size = 18)
p2 = plt.yticks(size = 18)
plt.legend(fontsize = 18)
```
