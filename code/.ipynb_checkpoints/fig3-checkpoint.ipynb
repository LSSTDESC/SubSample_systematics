```python
from pyccl import Cosmology as cosmologyccl
import numpy as np
Omega0_cdm=0.26377065934278865
Omega0_b=0.0482754208891869
h=0.67556
sigma8=0.8225
n_s=0.9667
redshift = 0.0
delta_crit = 1.686      # critical overdensity for collapse from Press-Schechter
c = 299792 #km/s
cosmoccl = cosmologyccl(Omega_c = Omega0_cdm, Omega_b = Omega0_b, h = h, sigma8 = sigma8, n_s = n_s)
H0 = cosmoccl['H0']
DH = (H0/c) # in units of h/Mpc, eq. 4 https://arxiv.org/pdf/astro-ph/9905116.pdf
Omega0_m = Omega0_cdm+Omega0_b
k_ = np.logspace(-4, 10, 10000)
Dc = cosmologyccl.comoving_radial_distance
Dlin = cosmoccl.growth_factor  
```

```python
import camb
from scipy.interpolate import InterpolatedUnivariateSpline as IUS
default_sigma8 = 0.79636317
default_As = 2e-9
P_k_max=10.0
scale = (sigma8/default_sigma8/1.0005116352583587)**2
As = scale*default_As
pars = camb.set_params(H0=100*h, ombh2=Omega0_b*h*h, omch2=Omega0_cdm*h*h, mnu=0.06, omk=0, tau=0.06,
                    ns=n_s, halofit_version='mead', lmax=3000,WantTransfer=True)
pars.InitPower.set_params(ns=n_s, As  = As)
pars.set_matter_power(redshifts=[0], kmax=P_k_max)
results= camb.get_results(pars)
```

```python
from camb import model, initialpower, CAMBparams, get_results
```

```python
kh,zs,PK = results.get_matter_power_spectrum(maxkh=10.0)
Plin_ = IUS(kh, PK[0,:], ext=1)
Plin = IUS(k_, Plin_(k_), ext=1)

trans = results.get_matter_transfer_data()
kh = trans.transfer_data[0,:,0]
trans_ = trans.transfer_data[model.Transfer_cdm-1,:,0]
Tlin_ = IUS(kh, trans_/trans_.max(), ext=1)
tk_ = Tlin_(k_)
is_good = np.isfinite(tk_)
Tlin = IUS(k_[is_good], tk_[is_good], ext=1)

```

```python
kh = np.logspace(-3,-1)
#plt.loglog(kh, Plin(kh))

alpha_fnl = 3.0*delta_crit*Omega0_m*(DH*DH)
delta_b = alpha_fnl/(kh**2*Tlin(kh)*Dlin(a = 0.8))
fnl = 10
fnl2 = 5
fnl3 = 0
delta_b = (1+fnl*alpha_fnl/(kh**2*Tlin(kh)*Dlin(a=0.8)))
delta_b2 = (1+fnl2*alpha_fnl/(kh**2*Tlin(kh)*Dlin(a=0.8)))
delta_b3 = (1+fnl3*alpha_fnl/(kh**2*Tlin(kh)*Dlin(a=0.8)))
plt.loglog(kh, delta_b*Plin(kh), label = "fnl = 10")
plt.loglog(kh, delta_b2*Plin(kh), label = "fnl = 5")
plt.loglog(kh, delta_b3*Plin(kh), label = "fnl = 0")
plt.legend()
```

```python
plt.figure(figsize=(7, 5))
plt.semilogx(kh, delta_b, label=r"$f_{nl} = 10$", linewidth=3)
plt.semilogx(kh, delta_b2, label=r"$f_{nl} = 5$", linewidth=3)


# Increase label sizes
plt.xlabel(r'$k\,(h^{-1} \mathrm{Mpc}^{-1})$', fontsize=18)
plt.ylabel(r'$b(k)$', fontsize=18)

# Increase tick label sizes
plt.tick_params(axis='both', which='major', labelsize=15)
plt.tick_params(axis='both', which='minor', labelsize=15)

# Add shaded regions
plt.fill_between(kh, np.ones_like(kh)*1, np.ones_like(kh)*1.2, alpha=0.2, color='k')
plt.fill_between(kh, np.ones_like(kh)*1, np.ones_like(kh)*1.1, alpha=0.2, color='k')
plt.fill_between(kh, np.ones_like(kh)*1, np.ones_like(kh)*1.02, alpha=0.2, color='k')


plt.legend(fontsize = 22)
p=plt.xticks(size = 18)
p=plt.yticks(size = 18)
plt.show()
```
